```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/experiment_results.csv")

df["engagement"] = (
    df["likes"]
    + df["replies"]
    + df["reposts"]
    + df["bookmarks"]
)

df.head()
```

```python
A = df[df["variant"] == "A"]["engagement"].values
B = df[df["variant"] == "B"]["engagement"].values
```

```python
A_mean = A.mean()
B_mean = B.mean()

A_std = A.std()
B_std = B.std()

A_mean, B_mean
```

```python
P(B > A)
```

```python
simulations = 50000

A_samples = np.random.normal(A_mean, A_std, simulations)
B_samples = np.random.normal(B_mean, B_std, simulations)

prob_B_better = (B_samples > A_samples).mean()

print("Probability B is better:", prob_B_better)
```

91% chance B is better

```python
lift = (B_mean - A_mean) / A_mean

print("Expected engagement lift:", lift)
```

```python
import matplotlib.pyplot as plt

plt.hist(A_samples, alpha=0.5, label="A")
plt.hist(B_samples, alpha=0.5, label="B")

plt.legend()
plt.title("Posterior Engagement Distributions")
plt.show()
```
