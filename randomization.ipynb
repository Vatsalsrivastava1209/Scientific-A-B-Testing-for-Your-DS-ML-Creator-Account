## Experiment Setup 

Variants: A vs B 

Tweets per variant: 17

Total tweets: 34

Posting frequency: 2 tweets/day

Duration: ~17 days

Metric: engagements per tweet

## Metric formula:

engagements = likes + replies + reposts + bookmarks

## Primary evaluation:

Mean engagements per tweet

```python
import random
import pandas as pd

n_per_variant = 17

variants = ["A"] * n_per_variant + ["B"] * n_per_variant
random.shuffle(variants)

df = pd.DataFrame({
    "tweet_number": range(1, len(variants) + 1),
    "variant": variants
})

df
```

```python
df.to_csv("tweet_schedule.csv", index=False)
```
